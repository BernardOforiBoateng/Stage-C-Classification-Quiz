```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/MaAsaCK/Desktop/Hamoye Internship/Classification/Stage C Dataset/Data_for_UCI_named.csv")
```

```python
df.head()
```

```python
#check distribution of target variable
df['stabf'].value_counts()
```

```python
#checking for null values in the dataset
df.isna().sum()
```

```python
df = df.drop(columns = ["stab"], axis  = 1)
```

```python
df.head()
```

```python
#check the datatype of the dataset
df.dtypes
```

```python
#One of the feature is categorical, so we need to encode it ahead 
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df.stabf = encoder.fit_transform(df.stabf)
```

```python
#let's preview the encoded feature
df.stabf
```

```python
#more preprocessing
X = df.drop(columns='stabf')
y = df['stabf']
```

```python
#split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
y_train.value_counts()
```

```python
#standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
normalised_train_df = scaler.fit_transform(X_train, y_train)
normalised_train_df = pd.DataFrame(normalised_train_df, columns = X_train.columns)
normalised_train_df
```

```python
#x_test = x_test.reset_index(drop=True)
normalised_test_df = scaler.transform(X_test)
normalised_test_df = pd.DataFrame(normalised_test_df, columns = X_test.columns)
normalised_test_df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
random_forest_y_pred = random_forest.predict(X_test)

random_forest_y_pred

print("Accuracy of Model::",accuracy_score(y_test,random_forest_y_pred))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
extra_trees = ExtraTreesClassifier(n_estimators=100)
extra_trees.fit(X_train, y_train)
extra_trees_y_pred = extra_trees.predict(X_test)

extra_trees_y_pred

print("Accuracy of Model::",accuracy_score(y_test,extra_trees_y_pred))
```
